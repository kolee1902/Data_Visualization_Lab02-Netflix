Để khám phá sự phân bố của các cột, chúng ta sẽ làm theo các bước sau:

- Cột số:
     - Tính toán và giải thích **số liệu thống kê tóm tắt** của các cột

     - Tính số giá trị còn thiếu trên mỗi cột, mỗi hàng và xử lý chúng

Đối với các cột có kiểu dữ liệu không phải là số, ta sẽ tính:
- Tỷ lệ giá trị bị thiếu [`Missing_ratio`]
- Số lượng giá trị duy nhất trong mỗi cột [`Num_diff_vals`]
- Phân phối từng cột (số lần xuất hiện của từng giá trị duy nhất) [`Distribution`]

```python
str_col_name = movie_df.select_dtypes(include=['object']).columns.tolist()

str_col_dist_df = pd.DataFrame(columns=str_col_name, 
                               index=['Missing_ratio', 'Num_diff_vals', 'Distribution'], 
                               dtype=object)

for column in str_col_name:
    str_col_dist_df.at['Missing_ratio', column] = movie_df[column].isnull().mean()
    str_col_dist_df.at['Num_diff_vals', column] = movie_df[column].nunique()
    str_col_dist_df.at['Distribution', column] = movie_df[column].value_counts(dropna=False).to_dict()

str_col_dist_df
```

```python
def count_occurrences(arr, df):
    occurrences = {}
    for string in arr:
        string_upper = string.upper()
        if string_upper in occurrences:
            occurrences[string_upper][0] += 1
        else:
            occurrences[string_upper] = [1, float('inf')]  
    for index, row in df.iterrows():
        name = row['Name']
        year = row['Year']
        if ':' in name:
            name = name.split(':')[0].strip().upper()
            if name in occurrences:
                occurrences[name][1] = min(occurrences[name][1], year)
    return occurrences

list_name = [name.split(':')[0].strip() for name in movie_df['Name'] if ':' in name]
res = count_occurrences(list_name, movie_df)

movie_dict = {name: [count, min_year] for name, [count, min_year] in res.items()}
```

```python
df = pd.DataFrame.from_dict(movie_dict, orient='index', columns=['Số mùa', 'Năm phát hành đầu tiên'])
df = df.reset_index()
df.columns = ['Tên phim', 'Số mùa', 'Năm phát hành đầu tiên']  # Đổi tên cột index thành 'Name'
df.sort_values(by='Số mùa', ascending=False).head(10)
```

```python
# Lấy danh sách tên phim, số lần xuất hiện và năm phát hành từ top 10 bộ phim
top_10_movies = sorted(movie_dict.items(), key=lambda x: x[1][0], reverse=True)[:10]
movies = [movie for movie, _ in top_10_movies]
counts = [count for _, [count, min_year] in top_10_movies]
years = [min_year for _, [_, min_year] in top_10_movies]

# Tạo subplot
color = ["#11264e", "#00507A", "#026e90", "#008b99", "#6faea4", "#fcdcb0", "#FEE08B", "#faa96e", "#f36b3b", "#ef3f28", "#CC0028"]
fig, ax = plt.subplots(figsize=(25, 10))

# Vẽ biểu đồ cột
bars = ax.barh(movies, counts, color=color, linewidth=2, edgecolor='k')

# Thêm thông tin về năm phát hành vào biểu đồ bằng cách sử dụng kích thước của các cột
for i in range(len(bars)):
    bar = bars[i]
    year = years[i]
    bar.set_width(counts[i])
    bar.set_label(f'{year}')

ax.set_xlabel('Numbers of Season', fontsize = 15)
ax.set_ylabel('Name of movie', fontsize = 15)
ax.set_title('Top 10 movies with the most Seasons and their first year of release on Netflix', fontsize = 20, weight='bold')
ax.invert_yaxis() 
ax.legend(fontsize='xx-large')
ax.grid(color='#95a5a6', linestyle='--', linewidth=1, axis='x', alpha=0.4)
plt.show()
```
